```python
import random
import nltk
from nltk.chat.util import Chat, reflections

# Define pairs of patterns and responses for the chatbot
pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, how can I help you today?",]
    ],
    [
        r"what is your name?",
        ["My name is Haster and I'm here to assist you.",]
    ],
    [
        r"how are you ?",
        ["I'm doing well, thank you!", "I'm fine, thank you!",]
    ],
    [
        r"sorry (.*)",
        ["It's alright", "No problem",]
    ],
    [
        r"(.*) (hungry|sleepy|bored|tired)",
        ["I can understand.", "I feel the same sometimes.",]
    ],
    [
        r"quit",
        ["Bye! Take care. See you soon.", "Have a great day!"]
    ],
    [
        r"(.*)",
        ["That's interesting, please tell me more.", "I see, can you elaborate?"]
    ]
]

# Create a chatbot instance
def chatbot():
    print("Hi, I'm Haster! How can I assist you today?")

    # Create Chat instance and pass pairs and reflections
    chat = Chat(pairs, reflections)

    # Start conversation
    while True:
        user_input = input("You: ")
        response = chat.respond(user_input)
        print("Haster:", response)
        if user_input.lower() == 'quit':
            break

if __name__ == "__main__":
    chatbot()
```
